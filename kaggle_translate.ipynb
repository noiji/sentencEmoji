```python
kakao_trans_api_token = '640f4d5ee381c90bc058d6d9491a1769'
```

```python
def send_kakao_trans_api(query):
    url = 'https://dapi.kakao.com/v2/translation/translate'
    headers = {"Authorization": "KakaoAK {}".format(kakao_trans_api_token)}
    params = {
        'src_lang': 'en',
        'target_lang': 'kr',
        'query': query
    }
    response = requests.get(url, headers = headers, params = params)
    print(response.json())
    ret = response.json()['translated_text'][0][0]
    return ret       
```

```python
kaggle_translated = open("data/kaggle_translated_restapi.txt", "r", encoding = "utf-8")
text_merged_api = kaggle_translated.read()
kaggle_translated.close()

count = 0

for raw in tqdm(text_en[864 + 834:]):
    t = send_kakao_trans_api(raw)
    text_merged_api += (t + raw[-2:-1] + ' ')
    
    if count < 5:
        print(t + raw[-2:-1])
    count += 1
    if count % 100 == 0:
        if count % 10000 == 0:
            print(t)
        kaggle_translated = open("data/kaggle_translated_restapi.txt", "w", encoding = "utf-8")
        kaggle_translated.write(text_merged_api)
        kaggle_translated.close()

text_merged_api[:500]
```
